```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 1 -> Create two 3Ã—3 matrices with random integers.

R = np.random.randint(20,30,(3, 3))
K = np.random.randint(30,40,(3,3))

# Task 2 -> Perform addition, subtraction, and multiplication operations on these matrices.

A=R+K
S=R-K
M=np.dot(R,K)

# Print all values

print(A)
print(S)
print(M)


# Task 3 -> Compute the determinant, inverse, and rank of one of the matrices.

det_R = np.linalg.det(R)

if det_R!=0:
    inv_R = np.linalg.inv(R)
else:
    inv_R = "Matrix is singular, cannot compute inverse"

rank_R = np.linalg.matrix_rank(R)

print(det_R)
print(inv_R)
print(rank_R)
```
