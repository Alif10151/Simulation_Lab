```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.rand(15)
y = np.random.rand(15)

print(x)
print(y)

# Create an index for x-axis

index = np.arange(len(x))
print(index)

# Plot the data

plt.plot(index, x, marker='o', linestyle='-', label="Vector X")
plt.plot(index, y, marker='s', linestyle='--', label="Vector Y")

# Labels and title

plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Two Random Matrices X and Y")
plt.legend()
plt.grid(True)

# Show the plot

plt.show()
```
